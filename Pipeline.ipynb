```python
# Load dataset
file_path = "Telco-Customer-Churn.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# drop customer id
df.drop(columns=['customerID'], inplace= True)
df.head()
```

```python
# Convert TotalCharges to numeric, handling errors
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce')
df.dropna(inplace=True)
```

```python
# Define features and target
target = "Churn"
categorical_features = [col for col in df.columns if df[col].dtype == 'object' and col != target]
numeric_features = ["SeniorCitizen", "tenure", "MonthlyCharges", "TotalCharges"]

print(f' category {categorical_features}')
print(f' numeric {numeric_features}')
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
OHE = OneHotEncoder(handle_unknown='ignore',
                    sparse_output=False,
                   ).set_output(transform='pandas')

OHE_df = OHE.fit_transform(df[['MultipleLines','InternetService', 'OnlineSecurity', 'OnlineBackup', 'TechSupport', 
                               'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']])
```

```python
OHE_df
```

```python
df2 = pd.concat( [df,OHE_df], axis=1 )
df2.head()
```

```python
df2.info()
```

```python
df2.drop(columns=['MultipleLines', 'InternetService', 'OnlineSecurity', 
                  'OnlineBackup', 'DeviceProtection', 'TechSupport', 
                  'StreamingTV', 'StreamingMovies', 'Contract', 
                  'PaperlessBilling', 'PaymentMethod'], inplace=True)
```

```python
# encode target 
df2['Churn'] = df2['Churn'].map({"Yes": 1, "No": 0})
```

```python
df2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df2.drop(columns=['Churn']), df['Churn'], test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
# Evaluate the model
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```
